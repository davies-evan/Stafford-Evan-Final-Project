```r
library(stringr)

library(dplyr)

library(ggplot2)

library(lubridate)

eleven_df <- read.csv("2011.csv")
```

```r
eleven_df <- filter(eleven_df, eleven_df$city != " ")
```

```r
eleven_df <- filter(eleven_df, eleven_df$race != " ")
```

```r
quantconvert <- function(v) {
    n <- length(v)
    column <- rep(0, n)
    
    for (i in 1:n) {
        if (v[i] == "Y") {
            column[i] <- 1
        }
        else if (v[i] == "N"){
            column[i] <- 0
        }
    }
    return(column)
}

eleven_df$arstmade_num <- quantconvert(eleven_df$arstmade)
```

```r
eleven_df$arstmade_num <- factor(eleven_df$arstmade_num)

race_arstmade3 <- glm(arstmade_num ~ race, data = eleven_df, family = "binomial")

summary(race_arstmade3)
```

```r
eleven_df$arstmade_num <- factor(eleven_df$arstmade_num)

city_arstmade3 <- glm(arstmade_num ~ city, data = eleven_df, family = "binomial")

summary(city_arstmade3)
```

```r
eleven_df_uniq <- unique(eleven_df[, c("datestop", "timestop", "crimsusp", "sex", "race", "age", "city", "arstmade", "arstmade_num")])
```

```r
eleven_df_uniq$arstmade_num <- factor(eleven_df_uniq$arstmade_num)

city_arstmade4 <- glm(arstmade_num ~ city, data = eleven_df_uniq, family = "binomial")

summary(city_arstmade4)
```

```r
eleven_df_uniq$arstmade_num <- factor(eleven_df_uniq$arstmade_num)

cityrace_arstmade <- glm(arstmade_num ~ city + race, data = eleven_df_uniq, family = "binomial")

summary(cityrace_arstmade)
```

```r
eleven_df_uniq$arstmade_num <- factor(eleven_df_uniq$arstmade_num)

cityraceint_arstmade <- glm(arstmade_num ~ city + race + city*race, data = eleven_df_uniq, family = "binomial")

summary(cityraceint_arstmade)
```
